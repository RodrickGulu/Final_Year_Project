```python
import os
import cv2
from tqdm import tqdm

def load_images_from_folders(folder_paths):
    images = []
    labels = []

    for folder_path in folder_paths:
        print(f"Loading images from folder: {folder_path}")
        
        # List all files in the folder
        file_list = os.listdir(folder_path)
        
        # Use tqdm to show a progress bar
        for filename in tqdm(file_list, desc=f"Processing {folder_path}", unit="file"):
            img_path = os.path.join(folder_path, filename)
            img = cv2.imread(img_path)
            if img is not None:
                images.append(img)
                labels.append(0)  # Assuming all images are labeled as 1 (you can adjust as needed)

    print(f"Total images loaded: {len(images)}")
    return images, labels

# Define paths to the retinal image folders
folder_paths = [
    '/kaggle/input/images-dataset/data/bike/',
    '/kaggle/input/images-dataset/data/cars/',
    '/kaggle/input/images-dataset/data/cats/',
    '/kaggle/input/images-dataset/data/dogs/',
    '/kaggle/input/images-dataset/data/flowers/',
    '/kaggle/input/images-dataset/data/horses/',
    '/kaggle/input/images-dataset/data/human/'
]

# Load retinal images from the specified folders
non_retinal_images, non_retinal_labels = load_images_from_folders(folder_paths)
```

```python
import numpy as np
import tensorflow as tf
import tensorflow_datasets as tfds
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Flatten
from tensorflow.keras.models import Model, Sequential
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import random

# Preprocess function to resize and normalize images
def preprocess_image(image, target_size=(224, 224)):
    image = tf.image.resize(image, target_size)
    image = tf.cast(image, tf.float32) / 255.0
    return image

non_retinal_images_resized = [preprocess_image(img) for img in non_retinal_images]
imagenet_images = np.array(non_retinal_images_resized)
non_retinal_labels = np.array(non_retinal_labels)

# Resize retinal images to match the size of imagenet images
retinal_images_resized = [preprocess_image(img) for img in retinal_images]
retinal_images_resized = np.array(retinal_images_resized)

# Concatenate retinal and non-retinal images and labels
images_combined = np.concatenate((retinal_images_resized, imagenet_images), axis=0)
labels_combined = np.concatenate((retinal_labels, non_retinal_labels), axis=0)

# Split dataset into train, validation, and test sets
train_images, test_images, train_labels, test_labels = train_test_split(images_combined, labels_combined, test_size=0.2, random_state=42)
train_images, val_images, train_labels, val_labels = train_test_split(train_images, train_labels, test_size=0.2, random_state=42)

# Load pre-trained ResNet50 model trained on ImageNet
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Extract features using ResNet50
def extract_features(images, model):
    features = model.predict(images)
    return features

# Extract features from train, validation, and test sets
train_features = extract_features(train_images, base_model)
val_features = extract_features(val_images, base_model)
test_features = extract_features(test_images, base_model)

# Flatten features
train_features = train_features.reshape((train_features.shape[0], -1))
val_features = val_features.reshape((val_features.shape[0], -1))
test_features = test_features.reshape((test_features.shape[0], -1))

print(train_features.shape)
print(val_features.shape)
print(test_features.shape)

# Create a fully connected neural network for classification
model = Sequential([
    Flatten(input_shape=train_features.shape[1:]),
    Dense(256, activation='relu'),
    Dense(1, activation='sigmoid')  # Binary classification (retinal vs non-retinal)
])

# Compile model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Plot 5 random images from the training set with their labels
def plot_images(images, labels):
    plt.figure(figsize=(10, 10))
    for i in range(5):
        idx = random.randint(0, len(images) - 1)
        plt.subplot(1, 5, i + 1)
        plt.imshow(images[idx])
        plt.title(f"Label: {labels[idx]}")
        plt.axis('off')
    plt.show()

plot_images(train_images, train_labels)

# Train model
history = model.fit(train_features, train_labels, epochs=10, validation_data=(val_features, val_labels))

# Evaluate model
test_loss, test_acc = model.evaluate(test_features, test_labels)
print('Test accuracy:', test_acc)
```

```python
import os
import cv2
from tqdm import tqdm

def load_images_from_folders(folder_paths):
    images = []
    labels = []

    for folder_path in folder_paths:
        print(f"Loading images from folder: {folder_path}")
        
        # List all files in the folder
        file_list = os.listdir(folder_path)
        
        # Use tqdm to show a progress bar
        for filename in tqdm(file_list, desc=f"Processing {folder_path}", unit="file"):
            img_path = os.path.join(folder_path, filename)
            img = cv2.imread(img_path)
            if img is not None:
                images.append(img)
                labels.append(1)  # Assuming all images are labeled as 1 (you can adjust as needed)

    print(f"Total images loaded: {len(images)}")
    return images, labels

# Define paths to the retinal image folders
folder_paths = [
    '/kaggle/input/g1020-final/G1020/glucoma'
]

# Load retinal images from the specified folders
retinal_images, retinal_labels = load_images_from_folders(folder_paths)
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
from sklearn.metrics import accuracy_score
from tqdm import tqdm

# Load pre-trained ResNet50 model trained on ImageNet (for feature extraction)
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Load the trained retinal model
model = load_model('/kaggle/working/retinal.h5')

# Preprocess function for a single image
def preprocess_image(image_input):
    if isinstance(image_input, str):
        img = image.load_img(image_input, target_size=(224, 224))
        img_array = image.img_to_array(img)
    else:
        img_array = tf.image.resize(image_input, (224, 224)).numpy()
    
    img_array = np.expand_dims(img_array, axis=0)
    img_array /= 255.0  # Normalize pixel values
    return img_array

# Function to extract features using the base model
def extract_features(images, model):
    features = model.predict(images, verbose=0)
    features = features.reshape((features.shape[0], -1))
    return features

# Prediction function for a list of images
def predict_images(images, labels):
    predictions = []
    feature = []
    for img in tqdm(images, desc="Processing images"):
        # Preprocess the image
        img_array = preprocess_image(img)
        
        # Extract features using the base model
        features = extract_features(img_array, base_model)
        
        # Predict using the trained model
        prediction = model.predict(features, verbose=0)
        
        # Append the prediction (1 for Retinal, 0 for Non-retinal)
        predictions.append(1 if prediction > 0.5 else 0)
        
        feature.append(features)
    
    # Calculate accuracy
    accuracy = accuracy_score(labels, predictions)
    
    return accuracy, predictions, feature

# Example usage
image_paths_or_arrays = retinal_images  # Combine retinal and non-retinal images
labels = retinal_labels  # Combine retinal and non-retinal labels

accuracy, predictions, feature = predict_images(image_paths_or_arrays, labels)
print(f'Accuracy: {accuracy}')
```
